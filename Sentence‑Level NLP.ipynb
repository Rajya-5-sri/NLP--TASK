```python
text = """Natural language processing (NLP) is a subfield of artificial intelligence, computer science, and linguistics. 
It focuses on the interactions between computers and human language; in particular, how to program computers to process 
and analyze large amounts of natural language data!

Applications of NLP are found in many areas: machine translation, sentiment analysis, speech recognition, and question-answering systems? 
These applications often rely on complex models, large datasets, and careful feature engineering."""
```

```python
import re
import nltk

from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

 ## Punctuation removal

```python

text_lower = text.lower()


text_no_punct = re.sub(r'[.,:;!?()]', ' ', text_lower)
text_no_punct = re.sub(r'\s+', ' ', text_no_punct).strip()

print("Original text:\n", text)
print("\nAfter lowercasing + punctuation removal:\n", text_no_punct)
```

## Tokenization and stop-word removal

```python
tokens = word_tokenize(text_no_punct)
stop_words = set(stopwords.words('english'))

tokens_no_sw = [w for w in tokens if w not in stop_words]

print("Tokens:", tokens)
print("After stop-word removal:", tokens_no_sw)
```

## Stemming

```python
stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(w) for w in tokens_no_sw]
print("After stemming:", stemmed_tokens)
```

## Lemmatization

```python
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(w) for w in tokens_no_sw]
print("After lemmatization:", lemmatized_tokens)
```

## BAG OF WORDS

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer


docs = [text]

cv = CountVectorizer()
bow_matrix = cv.fit_transform(docs)

print("Vocabulary:", cv.get_feature_names_out())
print("BoW matrix:\n", bow_matrix.toarray())
```

## TF-IDF


```python
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(docs)

print("TF-IDF vocabulary:", tfidf.get_feature_names_out())
print("TF-IDF matrix:\n", tfidf_matrix.toarray())
```

```python
from gensim.models import Word2Vec
from nltk.tokenize import sent_tokenize

nltk.download('punkt')

long_text = """Natural language processing (NLP) is a subfield of artificial intelligence, computer science, and linguistics. 
It focuses on the interactions between computers and human language; in particular, how to program computers to process 
and analyze large amounts of natural language data!

Applications of NLP are found in many areas: machine translation, sentiment analysis, speech recognition, and question-answering systems? 
These applications often rely on complex models, large datasets, and careful feature engineering."""

# tokenize into sentences, then words
sentences = sent_tokenize(long_text.lower())
sentences = [word_tokenize(s) for s in sentences]

# train Word2Vec
w2v_model = Word2Vec(
    sentences,
    vector_size=100,
    window=5,
    min_count=1,
    workers=4,
    sg=1  # skip-gram
)

# check vocabulary
print(list(w2v_model.wv.key_to_index.keys())[:30])
```

## Word similarity scores

```python
print("Similarity (nlp vs language):",
      w2v_model.wv.similarity('nlp', 'language'))

print("Similarity (intelligence vs data):",
      w2v_model.wv.similarity('intelligence', 'data'))

print("Most similar to 'language':",
      w2v_model.wv.most_similar('language'))
```

## Building sentence level corpus

```python
from nltk.tokenize import sent_tokenize


sent_list = sent_tokenize(text)
for i, s in enumerate(sent_list, 1):
    print(f"S{i}: {s}")
```

### Feature engineering

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import re
def clean_sent(s):
    s = s.lower()
    s = re.sub(r'[.,:;!?()]', ' ', s)
    s = re.sub(r'\s+', ' ', s).strip()
    return s

sent_clean = [clean_sent(s) for s in sent_list]

tfidf_sent = TfidfVectorizer()
X_sent = tfidf_sent.fit_transform(sent_clean)

print("No. of sentences:", len(sent_clean))
print("TF-IDF shape (sentences x features):", X_sent.shape)
```

### Cosine similarity between any 3 sentences

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# choose first 3 sentences
idx = [0, 1, 2]
X3 = X_sent[idx]

sim_matrix = cosine_similarity(X3.toarray())
print("Cosine similarity between sentences S1, S2, S3:")
print(np.round(sim_matrix, 3))

for i in idx:
    print(f"S{i+1}: {sent_list[i]}")
```

###  Supervised text classification 

```python
from sklearn.naive_bayes import MultinomialNB
y = [0, 0, 1, 1]   

clf = MultinomialNB()
clf.fit(X_sent, y)

y_pred = clf.predict(X_sent)
print("Supervised text classification (Naive Bayes):")
for i, (s, true, pred) in enumerate(zip(sent_list, y, y_pred), 1):
    print(f"S{i}  true={true}, pred={pred}  ->  {s}")
```

### Clustering sentences (unsupervised)

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2, random_state=42, n_init="auto")
clusters = kmeans.fit_predict(X_sent.toarray())

print("Unsupervised sentence clustering (KMeans):")
for i, (s, c) in enumerate(zip(sent_list, clusters), 1):
    print(f"S{i}  [Cluster {c}]  ->  {s}")
```
